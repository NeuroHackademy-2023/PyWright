```python
import numpy as np
import pandas as pd
import coneff
import load_data as ld
import tether
import nibabel as nib
import ml
```

```python
struc_mats_list = np.random.rand(70,70,1)
func_mats_list = np.random.rand(70,70,1)
bold_list = np.random.rand(70,200,1)
parcellation_dir = "../shared/HCP/115825/MNINonLinear/aparc+aseg.nii.gz"
meta_dir = "meta.csv"
output_labels = np.random.rand(1,1)
```

```python
parcellation = nib.load(parcellation_dir)
metadata = pd.read_csv(meta_dir)
```

```python
distance = np.load("tether_eucledian_matrix.npy")
```

```python
distance[:70, :70].shape
```

```python
# run tether
node_predictions, r_values = tether.tether(func_mats_list, struc_mats_list, parcellation, metadata)
```

```python
# run coneff
N = struc_mats_list.shape[0]
numSubj = struc_mats_list.shape[2]
effcon_direct = np.zeros([N,N,numSubj])
effcon_1degreeIndirect = np.zeros([N,N,numSubj])
for i in range(struc_mats_list.shape[2]):
    eff_conn, M_2nstep, etem, etem_2 = coneff.structured_G_causality(struc_mats_list[:,:,i], bold_list[:,:,i], use_multistep=1, maxiter=500)
    effcon_direct[:,:,i] = eff_conn
    effcon_1degreeIndirect[:,:,i] = eff_conn + M_2nstep
```

```python
# ML stuff
tether_ml_results = ml(r_values, output_labels)
coneff_ml_results = ml(effcon_1degreeIndirect, output_labels)
```
